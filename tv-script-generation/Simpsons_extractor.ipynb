```python
import pandas as pd
```

```python
df = pd.read_csv('./data/simpsons/simpsons_script_lines.csv', error_bad_lines=False, index_col='id')
df.head()
```

```python
text = df[(df['raw_location_text'] == 'Simpson Home') & (df['word_count'] > 0)] \
    .sort_values(['episode_id', 'timestamp_in_ms']) \
    .apply(lambda x: x.raw_character_text.replace(' ', '_') + ': ' + x.normalized_text, axis=1)
```

```python
text
```

```python
with open('./data/simpsons/simpsons_home.txt', "w") as text_file:
    for x in text:
        text_file.write(x + '\n')
```
